```python
import cv2
import numpy as np

# Load image
img = cv2.imread('flower.png', cv2.IMREAD_GRAYSCALE)

# Apply median filter with window size 5x5
median = cv2.medianBlur(img, 5)

# Apply Gaussian filter with standard deviation 1.5
gaussian = cv2.GaussianBlur(img, (5, 5), 1.5)

# Display filtered images side by side

plt.subplot(1, 2, 1)
plt.imshow(median)
plt.title('Median Filter')

plt.subplot(1, 2, 2)
plt.imshow(median)
plt.title('Gaussian Filter')

plt.show()
```

A median filter preserves edges and details while removing noise, while a Gaussian filter smooths the image using a weighted average and can result in loss of edge and detail information.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')
 
loaded_image = cv2.imread("flower.png")
loaded_image = cv2.cvtColor(loaded_image,cv2.COLOR_BGR2RGB)
 
gray_image = cv2.cvtColor(loaded_image,cv2.COLOR_BGR2GRAY)
 
edged_image1 = cv2.Canny(gray_image, threshold1=50, threshold2=150)
edged_image2 = cv2.Canny(gray_image, threshold1=100, threshold2=200)
edged_image3 = cv2.Canny(gray_image, threshold1=150, threshold2=250)
 
 
plt.figure(figsize=(20,20))
plt.subplot(1,3,1)
plt.imshow(loaded_image,cmap="gray")
plt.title("Original Image")
plt.axis("off")
plt.subplot(1,3,2)
plt.imshow(gray_image,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
```

```python
plt.subplot(1,3,1)
plt.imshow(edged_image1,cmap="gray")
plt.axis("off")
plt.title("Canny Edge Detected Img 1")
plt.subplot(1,3,2)
plt.imshow(edged_image2,cmap="gray")
plt.axis("off")
plt.title("Canny Edge Detected Img 2")
plt.subplot(1,3,3)
plt.imshow(edged_image3,cmap="gray")
plt.axis("off")
plt.title("Canny Edge Detected Img 3")
plt.show()
```

"A multistage algorithmic process for extracting edges from an image.
I. Noise reduction
II. Computation of Intensity Gradient
III. Non-maximum suppression of edges  - filters unwanted pixels
IV. Hysteresis thresholding"

The ease with which one can configure an edge detector to detect only objects is determined by the complexity of the image and the characteristics of the objects we want to detect. However, with the right tools and techniques, accurate object detection using edge detection algorithms is possible.
